Filtrado de algunos datos y selección de datos

Los datos los guarda como dataFrame que es como "tabla" y se van accediendo como listas pero es mucho más poderoso pues está sobre cargado el operador de listas.

Aqui se ve como regresa un vector con las evaluaciones de la operación > sobre los datos diciendo cuáles son mayores de 30

```python
# Filtering and Selecting
ageFilterValue = 30
ageFilter = data["Age"] > ageFilterValue
ageFilter
```

En este que sigue selecciona con el filtro ageFilter, los valores de data que tuvieron True para poder tener los registros que cumplen

```python
aboveAge = data[ageFilter]
aboveAge
```

```python
len(aboveAge)
```

unique() te permite saber cuáles son los valores que tiene esa columna del dataFrame

```python
data["Current Status"].unique()
```

Ahora vamos a obtener datos con más restricciones

```python
activeFilter = data["Current Status"] == "Active"

aboveAndActive = data[ageFilter & activeFilter]
len(aboveAndActive)
```
