Filtrado de algunos datos y selección de datos

Los datos los guarda como dataFrame que es como "tabla" y se van accediendo como listas pero es mucho más poderoso pues está sobre cargado el operador de listas.

Aqui se ve como regresa un vector con las evaluaciones de la operación > sobre los datos diciendo cuáles son mayores de 30

```python
# Filtering and Selecting
ageFilterValue = 30
ageFilter = data["Age"] > ageFilterValue
ageFilter
```

En este que sigue selecciona con el filtro ageFilter, los valores de data que tuvieron True para poder tener los registros que cumplen

```python
aboveAge = data[ageFilter]
aboveAge
```

```python
len(aboveAge)
```

unique() te permite saber cuáles son los valores que tiene esa columna del dataFrame

```python
data["Current Status"].unique()
```

Ahora vamos a obtener datos con más restricciones

```python
activeFilter = data["Current Status"] == "Active"

aboveAndActive = data[ageFilter & activeFilter]
len(aboveAndActive)
```

## Plotting

```python
aboveAndActive["Position"]
```

```python
frequencies = aboveAndActive["Position"].value_counts()
frequencies
```

```python
frequencies.plot.bar(
    title=(
        "Active players above " +
        str(ageFilterValue) +
        " grouped by position"
    )
)
```

# Plotting with style

```python
color = cm.inferno_r(np.linspace(.4, .8, 30))
frequencies.plot(
    kind="bar",
    title=(
        f"Active NFL players above {ageFilterValue} grouped by position"
    ),
    color=color
)
plt.savefig('./images/nflBar.png', dpi=500)
#plt.close() - se recomienda cerrar pero no todo mundo lo pone porque no despliega la gráfica
```

### Scatter plot

```python
sns.lmplot(
    x="Height (inches)", y="Weight (lbs)", data=aboveAndActive,
    aspect=1.5, fit_reg=False, hue='Position', legend=True,
    palette=sns.color_palette("hls", len(aboveAndActive["Position"].unique()))
)
plt.savefig('./images/nflScatter.png', dpi=500)
```
